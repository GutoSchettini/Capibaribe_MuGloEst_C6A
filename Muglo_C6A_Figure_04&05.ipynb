## Smoothing the data

```python
def filt_mtx(s):
    li, co = s.shape
    sf = np.full((li, co), np.nan)
    for l in range(li):
        i_good = np.where(~np.isnan(s[l,:]))[0]
        sf[l,i_good] = signal.filtfilt(b, a, s[l,i_good])
    return sf
```

```python
def calculations(t, x, v, filt=False):
    time = t*24*3600  # time in seconds 
    x = x*1000        # distance in meters
    n = v[0]
    u_s = v[1]
    u_b = v[2]
    rho_s = v[3]
    rho_b = v[4] 

    if filt == True:
        u_s = filt_mtx(u_s)
        u_b = filt_mtx(u_b)
        rho_s = filt_mtx(rho_s)
        rho_b = filt_mtx(rho_b)

    li, co = time.shape
    drho_dt_s = np.full((li, co), np.nan)
    drho_dt_b = np.full((li, co), np.nan)
    for l in range(li):
        drho_dt_s[l,:] = np.gradient(rho_s[l,:], time[l,:])
        drho_dt_b[l,:] = np.gradient(rho_b[l,:], time[l,:])
    
    u_drho_dx_s = np.full((li, co), np.nan)
    u_drho_dx_b = np.full((li, co), np.nan)
    for c in range(co):
        u_drho_dx_s[:,c] = -u_s[:,c] * np.gradient(rho_s[:,c], x[:,c])
        u_drho_dx_b[:,c] = -u_b[:,c] * np.gradient(rho_b[:,c], x[:,c])
    
    diffusive_s = drho_dt_s + u_drho_dx_s
    diffusive_b = drho_dt_b + u_drho_dx_b

    return drho_dt_s, drho_dt_b, u_drho_dx_s, u_drho_dx_b, diffusive_s, diffusive_b
    
```

```python
N_drdt_s, N_drdt_b, N_udrdx_s, N_udrdx_b, N_diff_s, N_diff_b = calculations(N_tt, N_xx, N_vi, filt=True)
S_drdt_s, S_drdt_b, S_udrdx_s, S_udrdx_b, S_diff_s, S_diff_b = calculations(S_tt, S_xx, S_vi, filt=True)

# put in the order for plotting
N_calc = [N_drdt_s, N_udrdx_s, N_diff_s, N_drdt_b,  N_udrdx_b,  N_diff_b]
S_calc = [S_drdt_s, S_udrdx_s, S_diff_s, S_drdt_b,  S_udrdx_b,  S_diff_b]
```

# Plotting

```python
# chatgpt - criando uma zona branca no zero do bwr

c_levels = 64

cmap = plt.get_cmap('bwr', c_levels)
colors = cmap(np.linspace(0, 1, c_levels))

f = 0.08
mid = int(c_levels*f)

center = c_levels // 2
colors[center-mid:center+mid] = [1,1,1,1]

new_cmap = mcolors.ListedColormap(colors, name='bwr2')
```

```python
letters = ['(a) /t $(kg/m^3/s)$', '(b) -u /x $(kg/m^3/s)$', '(c) D $(kg/m^3/s)$']
```

```python
fig = plt.figure(figsize=(12, 8))

px = .1
py = .8
dx = .3
dy = .18
pxcb1 = px+dx+.015

int_v = .04
int_h = .1

px2 = px+dx+int_h
pxcb2 = px2+dx+.015

ax_c1 = []
ax_c2 = []
ax_c1_cb = []
ax_c2_cb = []
for i in range(3):
    ax_c1.append(fig.add_axes([px, py-(dy+int_v)*i, dx, dy]))
    ax_c2.append(fig.add_axes([px2, py-(dy+int_v)*i, dx, dy]))

    ax_c1_cb.append(fig.add_axes([pxcb1, py-(dy+int_v)*i, .01, dy]))
    ax_c2_cb.append(fig.add_axes([pxcb2, py-(dy+int_v)*i, .01, dy]))

    norm = mcolors.TwoSlopeNorm(vmin = np.nanmin(N_calc[i]), vcenter=0, vmax =np.nanmax(N_calc[i]))
    cb1 = ax_c1[i].contourf(N_tt_dt, N_xx, N_calc[i], cmap=new_cmap, norm=norm)
    cbar1 = plt.colorbar(cb1, cax=ax_c1_cb[i])
   
    ax_c1[i].contour(N_tt_dt, N_xx, N_calc[i], [0], colors='k')

    norm = mcolors.TwoSlopeNorm(vmin = np.nanmin(S_calc[i]), vcenter=0, vmax =np.nanmax(S_calc[i]))
    cb2 = ax_c2[i].contourf(S_tt_dt, S_xx, S_calc[i], cmap=new_cmap, norm=norm)
    cbar2 = plt.colorbar(cb2, cax=ax_c2_cb[i])
    ax_c2[i].contour(S_tt_dt, S_xx, S_calc[i], [0], colors='k')

    cbs = [cbar1, cbar2]
    for cb in cbs:
        ticks = cb.get_ticks()
        select = ticks[::2]
        cb.set_ticks(select)
        cb.formatter.set_scientific(True)
        cb.formatter.set_powerlimits((0,0))
        
    axs = [ax_c1[i], ax_c2[i]]
    axs[0].set_ylabel('Dist. (km)')
    axs[1].set_xlim(S_tt_dt[0,0], S_tt_dt[0,-29])
    for j, ax in enumerate(axs):
        ax.xaxis.set_major_locator(mdates.HourLocator(interval=6))
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
        if i < 2:
            ax.set_xticklabels('')
        else:
            if j == 0:    
                ax.set_xlabel('21-22 March')
            else:
                ax.set_xlabel('26-27 March')

    ax_c1[i].text(0, 1.02, letters[i], transform = ax_c1[i].transAxes, fontweight='bold')

fig.text(.22, 1.03, 'Neap tide', fontsize=16, fontweight='bold')
fig.text(.6, 1.03, 'Spring tide', fontsize=16, fontweight='bold')
fig.text(.41, 1.03, '(surface)', fontsize=14, fontweight='bold')
```
