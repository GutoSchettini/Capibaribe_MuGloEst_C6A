# Calculations... 2

```python
def filt_mtx(s):
    li, co = s.shape
    sf = np.full((li, co), np.nan)
    for l in range(li):
        i_good = np.where(~np.isnan(s[l,:]))[0]
        sf[l,i_good] = signal.filtfilt(b, a, s[l,i_good])
    return sf
```

```python
def calculations_2(t, x, v, filt=False):
    time = t*24*3600  # time in seconds 
    x = x*1000        # distance in meters
    z = v[0]          # depth
    u_s = v[1]
    u_b = v[2]
    rho_s = v[3]
    rho_b = v[4] 

    if filt == True:
        u_s = filt_mtx(u_s)
        u_b = filt_mtx(u_b)
        rho_s = filt_mtx(rho_s)
        rho_b = filt_mtx(rho_b)
    
    drho_dz = (rho_s - rho_b)/z
    
    li, co = time.shape
    
    # stratification changes
    ddt_drho_dz = np.full((li,co), np.nan)
    for l in range(li):
        ddt_drho_dz[l,:] = np.gradient(drho_dz[l,:], time[l,:])
    
    stratification_changes = ddt_drho_dz
    
    # tidal straining
    rho_da = (rho_s + rho_b)/2  # depth averaged < > 
    
    drho_da_dx = np.full((li,co), np.nan)
    for c in range(co):
        drho_da_dx[:,c] = np.gradient(rho_da[:,c], x[:,c])
    
    du_dz = (u_s - u_b)/z
    
    tidal_straining = du_dz * drho_da_dx
    
    # advection of gradient
    u_da = (u_s + u_b)/2   # or we should use all vertical data? but here we are 
    
    ddx_drho_dz = np.full((li,co), np.nan)
    for c in range(co):
        ddx_drho_dz[:,c] = np.gradient(drho_dz[:,c], x[:,c])
    
    advection_gradient = u_da * ddx_drho_dz
    
    mixing = stratification_changes - tidal_straining - advection_gradient

    return stratification_changes, -tidal_straining, -advection_gradient, mixing  # inverting the signal to fit with Arnoldo's notation
```

```python
N_strat, N_strain, N_advect, N_mixing = calculations_2(N_tt, N_xx, N_vi, filt=True)
S_strat, S_strain, S_advect, S_mixing = calculations_2(S_tt, S_xx, S_vi, filt=True)

N_calc_2 = [N_strat, N_strain, N_advect, N_mixing]
S_calc_2 = [S_strat, S_strain, S_advect, S_mixing]
```
